```python
import pandas as pd
import numpy as np
```

```python
#inputpd = pd.read_csv("List_Example.csv")
data=[["T001","F;A;E;G"],["T002","B"],["T003","F"],["T004","G;C"],["T005","C;D"],["T006","F;B;D"],["T007","A"],
      ["T008","G;A"],["T009","D;G;A"],["T010","F;G"],["T011","G;C"],["T012","F;C;A"],["T013","D"],["T014","F;C"],
      ["T015","F"],["T016","E"],["T017","D;F"],["T018","A"],["T019","F"],["T020","A;F;D"],["T021","E"],
      ["T022","C;F;F"],["T023","C"],["T024","D"],["T025","A"],["T026","G"],["T027","F"],["T028","C"],
      ["T029","G"],["T030","B"],["T031","F"],["T032","B;G;D"],["T033","E;G"],["T034","E"],["T035","D"],
      ["T036","D;C"],["T037","B"],["T038","D"],["T039","D;C"],["T040","C"]]
inputpd=pd.DataFrame(data,columns=['TID','TType'])
```

```python
datType=inputpd["TType"]
datType2=datType.str.split(";",-1,False)
```

```python
alist=[a for row in datType2 for a in row]
nplist=np.array(alist)
```

```python
uni,ct=np.unique(nplist,return_counts=True)
dictype=dict(zip(uni,ct))
```

```python
def getCts(types):
    olst=types.split(";")
    allv=list()
    for o in olst:
        allv.append(dictype[o])
    s = "".join(str(allv))
    return s.replace(",",";").replace("[","").replace("]","")
    #rslt=np.array(allv)
    #return np.unique(rslt)
def getGroups(cts):
    alst=cts.split(";")
    allv=list()
    for ct in alst:
        if int(ct)<=5: 
            allv.append("Group-A (01-05)")
        else:
            if int(ct)>10:
                allv.append("Group-C (10+)")
            else:
                allv.append("Group-B (06-10)")
    rslt=np.array(allv)
    rslt=np.unique(rslt)
    s = ",".join(rslt)
    return s.replace(",",";").replace("[","").replace("]","")    
```

```python
inputpd["TypeCts"]=[getCts(row) for row in inputpd["TType"]]
inputpd["TypeGroup"]=[getGroups(row) for row in inputpd["TypeCts"]]
inputpd
```
